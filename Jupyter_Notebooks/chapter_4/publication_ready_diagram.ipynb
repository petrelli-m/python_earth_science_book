# Prepare a publication-ready diagram

<i>This is about code listing 4.7</i><br/>
...as a final task, we prepare a publication-ready diagram: 
<ul>
    <li>Lines 1 and 2 import the pandas library and the <i>matplotlib.pyplot</i> module, respectively;</li>
    <li>Line 4 imports the Excel file (Smith et al., 2011) in a DataFrame named <i>my_dataset</i>;</li> 
    <li>Lines 7-9 define three sequences named epochs, colors, and markers, respectively;</li>
    <li>At line 11, we generate a new empty figure, and then line 12 implements a loop that iterates over the sequences of epochs, colors, and markers using the <i>zip()</i> function. 
    <li>This enables us to iterate over two or more lists at the same time.</li> 
    <li>Next, line 13 defines a new DataFrame named <i>my_data</i> by filtering <i>my_dataset</i> using the labels in the Epoch column.</li> 
    <li>At line 14, we add a Zr versus Th scatter diagram of the resulting <i>my_data</i> to the figure generated at line 11.</li>
    <li>Finally, we add axis labels (lines 16 and 17) and a legend with a title (line 18). </li>
    <li>Note that  \n simply defines a new line in the legend title.</li>

```python
import pandas as pd
import matplotlib.pyplot as plt

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')


epochs = ['one', 'two', 'three', 'three-b']
colors = ['#c8b4ba', '#f3ddb3', '#c1cd97', '#e18d96']
markers = ['o', 's', 'd', 'v']

fig, ax = plt.subplots(figsize=(15,10))
for (epoch, color, marker) in zip(epochs, colors, markers):
    my_data = my_dataset[(my_dataset.Epoch == epoch)]
    ax.scatter(my_data.Zr, my_data.Th, marker=marker, s=100, c=color, edgecolor='0', label="Epoch " + epoch)

ax.set_xlabel("Zr [ppm]")
ax.set_ylabel("Th [ppm]")
ax.legend(title="Phlegraean Fields \n Age < 15 ky")
```

We can also replicate Figs. 4d anf 4e by Smith et al. (2011):

```python
epochs = ['one', 'two', 'three', 'three-b']
# Using an arbitrary color palette
colors = ['#c8b4ba', '#f3ddb3', '#c1cd97', '#e18d96']
markers = ['o', 's', 'd', 'v']

fig, (ax1, ax2) = plt.subplots(1,2,figsize=(15,5))

for (epoch, color, marker) in zip(epochs, colors, markers):
    my_data = my_dataset[(my_dataset.Epoch == epoch)]
    ax1.scatter(my_data.Zr, my_data.Ba, marker=marker, s=80, c=color, edgecolor='0', label="Epoch " + epoch)
    ax2.scatter(my_data.Rb, my_data.Sr, marker=marker, s=80, c=color, edgecolor='0', label="Epoch " + epoch)

ax1.set_title("Fig 4d (Smith et al., 2011)")
ax1.set_xlabel("Zr [ppm]")
ax1.set_ylabel("Ba [ppm]")
ax2.set_title("Fig 4e (Smith et al., 2011)")
ax2.set_xlabel("Rb [ppm]")
ax2.set_ylabel("Sr [ppm]")


ax1.legend(title="Phlegraean Fields \n Age < 15 ky")
```

Note that "In individuals of Northern European ancestry, as many as 8 percent of men and 0.5 percent of women experience the common form of red-green color blindness. If a submitted manuscript happens to go to three male reviewers of Northern European descent, the chance that at least one will be color blind is 22 percent." (<a href="https://www.nature.com/articles/nmeth.1618">Wong, 2011</a>). So let start using colors optimized for color-blind individuals:

```python
epochs = ['one', 'two', 'three', 'three-b']

# Using a palette for color-blind individuals from Wong, 2011
colors = ['#e69f00', '#56b3e9', '#f0e442', '#cc79a7']
markers = ['o', 's', 'd', 'v']

fig, (ax1, ax2) = plt.subplots(1,2,figsize=(15,5))

for (epoch, color, marker) in zip(epochs, colors, markers):
    my_data = my_dataset[(my_dataset.Epoch == epoch)]
    ax1.scatter(my_data.Zr, my_data.Ba, marker=marker, s=80, c=color, edgecolor='0', label="Epoch " + epoch)
    ax2.scatter(my_data.Rb, my_data.Sr, marker=marker, s=80, c=color, edgecolor='0', label="Epoch " + epoch)

ax1.set_title("Fig 4d (Smith et al., 2011)")
ax1.set_xlabel("Zr [ppm]")
ax1.set_ylabel("Ba [ppm]")
ax2.set_title("Fig 4e (Smith et al., 2011)")
ax2.set_xlabel("Rb [ppm]")
ax2.set_ylabel("Sr [ppm]")


ax1.legend(title="Phlegraean Fields \n Age < 15 ky")
```
